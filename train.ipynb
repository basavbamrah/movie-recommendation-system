# Moview Recommendation System 

**Types  of Rocommendatioin Systems**

Machine Leaninig algorithims in recommender systems typicaaly fit into two categories:
1. Content-based systems
2. collaborative filtering systems

**1. Content-Based Movie Recommendation Systems**
<!-- <img src='test.jpg -->
Content-based methods are based on the similarity of movie attributes.Using this type of recommender system, is a user watches one movie, similar movies are recommended.
Example: if a user watches a moview starring *Tom Cruise* the system will recommenf them movies in the same genre starring *Tom Cruise*.

<img src='./content-based.jpg'>

**2. Collaborative Filtering Movie Recommendation Systems**

With collaborative filteringm the system is based in past intertacions between user and movies. The system will recommend movies that are similar to movies that a user has watched in the past.
Example: if a user **A** watches movie M1, M2 and M3 and user **B** watches M1, M2,M3 and M4, the system will recommend M4 to user **C**.
