# Collaborative Filtering

It has two parts:
1. User-based collaborative filtering:
    <p>
        These system recommend movies to a user that a similar user has liked. The measure of similarity is based on the ratings given by the user. The similarity is calculated using the cosine similarity, or the Pearson correlation coefficient. 
    </p>
2. Item-based collaborative filtering:
    <p>
        These system recommend movies to a user that are similar to the movies that the user has liked. The measure of similarity is based on the ratings given by the user. The similarity is calculated using the cosine similarity, or the Pearson correlation coefficient.
    </p>
